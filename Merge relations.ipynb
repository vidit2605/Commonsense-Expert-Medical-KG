## Merging the files with relations

```python
import pandas as pd
import glob
import os
  
# merging the files
joined_files = os.path.join("/home/akumar/", "relations-*.csv")
  
# A list of all joined files is returned
joined_list = glob.glob(joined_files)
  
# Finally, the files are joined
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
df
```

```python
#saving as csv
df.to_csv("/home/akumar/merged_relations.csv")
```

```python
#counting number of occurrences of each relation type
dict1={}
for val in df["relation"]:
    if val in dict1.keys():
        dict1[val]+=1
    else:
        dict1[val]=1
dict1
```

### Updating type of relation based on links

```python
df.loc[(df['relation'] == 'TrCP') & (df['entity1'] == 'PROBLEM'),'relation']= 'CAUSED BY'
df.loc[(df['relation'] == 'TrCP') & (df['entity1'] == 'TREATMENT'),'relation']= 'CAUSES PROBLEM'

df.loc[(df['relation'] == 'TrAP') & (df['entity1'] == 'PROBLEM'),'relation']= 'HAS TREATMENT'
df.loc[(df['relation'] == 'TrAP') & (df['entity1'] == 'TREATMENT'),'relation']= 'TREATS PROBLEM'

df.loc[(df['relation'] == 'TeRP') & (df['entity1'] == 'PROBLEM'),'relation']= 'DETECTED BY'
df.loc[(df['relation'] == 'TeRP') & (df['entity1'] == 'TEST'),'relation']= 'DETECTS'

df.loc[(df['relation'] == 'TeCP') & (df['entity1'] == 'PROBLEM'),'relation']= 'DETECTED BY'
df.loc[(df['relation'] == 'TeCP') & (df['entity1'] == 'TEST'),'relation']= 'DETECTS'

df.loc[(df['relation'] == 'TrIP') & (df['entity1'] == 'PROBLEM'),'relation']= 'CURED BY'
df.loc[(df['relation'] == 'TrIP') & (df['entity1'] == 'TREATMENT'),'relation']= 'CURES'

df.loc[(df['relation'] == 'TrNAP') & (df['entity1'] == 'PROBLEM'),'relation']= 'MUST AVOID'
df.loc[(df['relation'] == 'TrNAP') & (df['entity1'] == 'TREATMENT'),'relation']= 'AVOIDED FOR'

df.loc[(df['relation'] == 'TrWP') & (df['entity1'] == 'PROBLEM'),'relation']= 'WORSENED BY'
df.loc[(df['relation'] == 'TrWP') & (df['entity1'] == 'TREATMENT'),'relation']= 'WORSENS'
```

```python
values = ['TrCP','TrAP','TeRP','TeCP','TrIP','TrNAP','TrWP']

df_up = df[df.relation.isin(values) == False]
```

```python
#dropping rows with null values
df_up.dropna(inplace=True)
```

### Removing duplicates

```python
#making chunk1 lowercase
df_up["chunk1"] = df_up["chunk1"].str.lower()
```

```python
#making chunk2 lowercase
df_up["chunk2"] = df_up["chunk2"].str.lower()
```

```python
df_up
```

```python
#remove duplicates
df_final = df_up.drop_duplicates(subset=['chunk1','chunk2'])
```
